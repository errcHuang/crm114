##Parsing OTU file (.shared) into .lrn, .cls, and .names

So by examining the structure of the data files, I determined that 

n = # of samples (51)
m = # of OTUs (2741+1)

.lrn file

I wrote a script that calculates what percent of a given sample each OTU (row) would take. I emailed Jarrod, and he recommended that I throw out OTUs that have 3 or fewer samples, and throw out samples that have a number of reads that is way less than average(reads being the sum of all the OTU data). All these things can be calculated by summing the rows and columns. 

Basically toss out OTUs that are only found in 1/2/3 samples and toss out samples with less than 100 reads.

.cls file

For the .cls file, I divided up the samples based off of site location in the metadata file Jarrod sent (2. Supplementary Table 1 Master.txt) and just hardcoded the divisions.
